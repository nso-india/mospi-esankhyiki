# nsoindia - Examples

Python client for India's National Statistical Office (MoSPI) data portal.
Access indicators across 19 datasets: employment, prices, GDP, health, education, environment, and more.

pip install nsoindia[pandas]

````python
import warnings
warnings.filterwarnings('ignore')
import nsoindia
from nsoindia.exceptions import InvalidDatasetError, InvalidFilterError, APIError, NoDataError
````

---
## Available Datasets

````python
nsoindia.list_datasets(format="df")
````

---
## PLFS - Periodic Labour Force Survey

Covers labour force participation, unemployment, worker population ratio, wages.
Use `get_indicators` to see what indicator codes are available, `get_metadata` to see valid filter values for a given indicator.

````python
# Indicators grouped by frequency_code
indicators = nsoindia.get_indicators("PLFS")
for group, items in indicators["indicators_by_frequency"].items():
    print(f"\n{group}:")
    for item in items:
        print(f"  {item['indicator_code']}: {item.get('indicator_name', item.get('description', ''))}")
````

````python
# Valid filter values for Unemployment Rate (indicator_code=3), Annual (frequency_code=1)
metadata = nsoindia.get_metadata("PLFS", indicator_code=3, frequency_code=1,)
for key, values in metadata["filter_values"]["data"].items():
    print(f"\n{key}:")
    for v in values[:4]:
        print(f"  {v}")
````

````python
# Unemployment Rate - All India, all years, both sectors
nsoindia.get_data("PLFS", {
    "indicator_code": 3,
    "frequency_code": 1,
    "state_code": 99,     # 99 = All India
    "gender_code": 3,     # 3 = Person
    "age_code": 1,        # 1 = 15 years and above
    "sector_code": 3,     # 3 = rural + urban
}, format="df")
````

````python
# Average wages - All India, annual, 2023-24
nsoindia.get_data("PLFS", {
    "indicator_code": 2,   # WPR (Worker Population Ratio, in per cent)
    "frequency_code": 1,
    "year": "2023-24",
    "state_code": 99,
    "gender_code": 3,
    "age_code": 1,
    "sector_code": 3,
}, format="df")
````

## CPI - Consumer Price Index

Multiple base years (2010, 2012, 2024) and two levels: Group and Item.

````python
# Available base years, levels, and series
nsoindia.get_indicators("CPI", format="df")
````

````python
# Filter values for base_year=2012, Group level
metadata = nsoindia.get_metadata("CPI", base_year="2012", level="Group", series="Current", format="df")
metadata
````

````python
# CPI Group index, base year 2012, year 2024
nsoindia.get_data("CPI", {
    "base_year": "2012",
    "series": "Current",
    "year": "2024",
}, format="df")
````

## NAS - National Accounts Statistics (GDP)

Annual and quarterly GDP, GVA, consumption, capital formation by industry and institutional sector.

````python
nsoindia.get_indicators("NAS", format="df")
````

````python
# Filter values for GVA by industry (indicator_code=1), base year 2022-23
metadata = nsoindia.get_metadata("NAS", indicator_code=1, base_year="2022-23", series="Current", frequency_code=1)
for key, values in metadata.items():
    if isinstance(values, list):
        print(f"{key}: {values[:3]}")
````

````python
# GVA by industry, current prices, latest estimates
nsoindia.get_data("NAS", {
    "indicator_code": 1,
    "base_year": "2022-23",
    "series": "Current",
    "frequency_code": 1,
    "year": "2024-25",
}, format="df")
````

---
## Error Handling

````python
# Typo in dataset name - fuzzy suggestion
try:
    nsoindia.get_indicators("plfs_data")
except InvalidDatasetError as e:
    print(e)
````

````python
# Missing required parameter
try:
    nsoindia.get_metadata("PLFS", frequency_code=1)
except InvalidFilterError as e:
    print(e)
````

````python
# Invalid filter parameter name
try:
    nsoindia.get_data("PLFS", {
        "indicator_code": 1,
        "frequency_code": 1,
        "typo_param": "hello",
    })
except InvalidFilterError as e:
    print(e)
````
